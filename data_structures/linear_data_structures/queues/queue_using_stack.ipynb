## Uses 2 stacks, implemented using arrays
## SC: O(n) -> since total elements between the 2 stacks will always be n
## TC: O(1) [Amortized Analysis]

```python
class Queue:
    def __init__(self):
        self.push_stack = []
        self.pop_stack = []
        
    def enqueue(self, value):
        self.push_stack.append(value)
    
    def dequeue(self):
        if not self.pop_stack:
            while len(self.push_stack) != 1:
                popped_el = self.push_stack.pop()
                self.pop_stack.append(popped_el)
            self.push_stack.pop()
        else:
            self.pop_stack.pop()
        
    def peek(self):
        if not self.pop_stack and not self.push_stack:
            return False
        elif not self.pop_stack:
            return self.push_stack[0]
        else:
            return self.pop_stack[len(self.pop_stack)-1]
    
    def is_empty(self):
        if not self.push_stack and self.pop_stack:
            return True
        else: 
            return False
```

```python
queue = Queue()
print('Push Stack:', queue.push_stack)
print('Pop Stack:', queue.pop_stack)
print('------------------------------')
queue.enqueue(5)
print('Push Stack:', queue.push_stack)
print('Pop Stack:', queue.pop_stack)
print('------------------------------')
queue.enqueue(1)
print('Push Stack:', queue.push_stack)
print('Pop Stack:', queue.pop_stack)
print('------------------------------')
queue.enqueue(2)
print('Push Stack:', queue.push_stack)
print('Pop Stack:', queue.pop_stack)
print('------------------------------')
queue.dequeue()
print('Push Stack:', queue.push_stack)
print('Pop Stack:', queue.pop_stack)
print('------------------------------')
print(queue.peek())
print('------------------------------')
queue.enqueue(8)
print('Push Stack:', queue.push_stack)
print('Pop Stack:', queue.pop_stack)
print('------------------------------')
queue.enqueue(6)
print('Push Stack:', queue.push_stack)
print('Pop Stack:', queue.pop_stack)
print('------------------------------')
print(queue.peek())
print('------------------------------')
queue.dequeue()
print('Push Stack:', queue.push_stack)
print('Pop Stack:', queue.pop_stack)
print('------------------------------')
queue.dequeue()
print('Push Stack:', queue.push_stack)
print('Pop Stack:', queue.pop_stack)
print('------------------------------')
print(queue.peek())
print('------------------------------')
queue.dequeue()
print('Push Stack:', queue.push_stack)
print('Pop Stack:', queue.pop_stack)
print('------------------------------')
queue.dequeue()
print('Push Stack:', queue.push_stack)
print('Pop Stack:', queue.pop_stack)
print('------------------------------')
print(queue.peek())
print('------------------------------')
```
