```python
def print_tree(root):
    def height(root):
        return 1 + max(height(root.left), height(root.right)) if root else -1  
    nlevels = height(root)
    width =  pow(2,nlevels+1)

    q=[(root,0,width,'c')]
    levels=[]

    while(q):
        node,level,x,align= q.pop(0)
        if node:            
            if len(levels)<=level:
                levels.append([])
        
            levels[level].append([node,level,x,align])
            seg= width//(pow(2,level+1))
            q.append((node.left,level+1,x-seg,'l'))
            q.append((node.right,level+1,x+seg,'r'))

    for i,l in enumerate(levels):
        pre=0
        preline=0
        linestr=''
        pstr=''
        seg= width//(pow(2,i+1))
        for n in l:
            valstr= str(n[0].value)
            if n[3]=='r':
                linestr+=' '*(n[2]-preline-1-seg-seg//2)+ '¯'*(seg +seg//2)+'\\'
                preline = n[2] 
            if n[3]=='l':
               linestr+=' '*(n[2]-preline-1)+'/' + '¯'*(seg+seg//2)  
               preline = n[2] + seg + seg//2
            pstr+=' '*(n[2]-pre-len(valstr))+valstr #correct the potition acording to the number size
            pre = n[2]
        print(linestr)
        print(pstr)   
```

```python
class BST_Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None
        
class BinarySearchTree:
    
    def __init__(self):
        self.root = None
                    
    def insert(self, node, value):
        if not node:
            return BST_Node(value)
        # LEFT
        if node.value > value:
            node.left = self.insert(node.left, value)
        # RIGHT
        elif node.value < value:
            node.right = self.insert(node.right, value)
        return node
    
    # Return no. of nodes in BST
    def size(self, node):
        if not node:
            return 0
        left_nodes = self.size(node.left)
        right_nodes = self.size(node.right)
        total_nodes = left_nodes + right_nodes + 1
        return total_nodes
    
    # Return sum of nodes in BST
    def sum_nodes(self, node):
        if not node:
            return 0
        left_sum = self.sum_nodes(node.left)
        right_sum = self.sum_nodes(node.right)
        total_sum = left_sum + right_sum + node.value
        return total_sum
    
    # Return node with greatest value in BST
    def maxm(self, node):
        if not node.right:
            return node.value
        return self.maxm(node.right)
    
    # Return node with smallest value in BST
    def minm(self, node):
        if not node.left:
            return node.value
        return self.minm(node.left)
    
    def height(self, node):
        if not node:
            return -1 # Return -1 if height in terms of edges else return 0 if height in terms of nodes
        left_height = self.height(node.left)
        right_height = self.height(node.right)
        total_height = max(left_height, right_height) + 1
        return total_height
```
