```python
def print_tree(root):
    def height(root):
        return 1 + max(height(root.left), height(root.right)) if root else -1  
    nlevels = height(root)
    width =  pow(2,nlevels+1)

    q=[(root,0,width,'c')]
    levels=[]

    while(q):
        node,level,x,align= q.pop(0)
        if node:            
            if len(levels)<=level:
                levels.append([])
        
            levels[level].append([node,level,x,align])
            seg= width//(pow(2,level+1))
            q.append((node.left,level+1,x-seg,'l'))
            q.append((node.right,level+1,x+seg,'r'))

    for i,l in enumerate(levels):
        pre=0
        preline=0
        linestr=''
        pstr=''
        seg= width//(pow(2,i+1))
        for n in l:
            valstr= str(n[0].value)
            if n[3]=='r':
                linestr+=' '*(n[2]-preline-1-seg-seg//2)+ '¯'*(seg +seg//2)+'\\'
                preline = n[2] 
            if n[3]=='l':
               linestr+=' '*(n[2]-preline-1)+'/' + '¯'*(seg+seg//2)  
               preline = n[2] + seg + seg//2
            pstr+=' '*(n[2]-pre-len(valstr))+valstr #correct the potition acording to the number size
            pre = n[2]
        print(linestr)
        print(pstr)   
```

```python
class BST_Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None
        
class BinarySearchTree:
    
    def __init__(self):
        self.root = None
        
    def lookup(self, node, value):
        if not node:
            return None
        # LEFT
        if node.value > value:
            return self.lookup(node.left, value)
        # RIGHT
        elif node.value < value:
            return self.lookup(node.right, value)
        # MATCH FOUND
        else:
            return node
                    
    def insert(self, node, value):
        if not node:
            return BST_Node(value)
        # LEFT
        if node.value > value:
            node.left = self.insert(node.left, value)
        # RIGHT
        elif node.value < value:
            node.right = self.insert(node.right, value)
        return node
        
    def remove(self, node, value):
        if not node:
            return None
        # LEFT
        if node.value > value:
            node.left = self.remove(node.left, value);
        # RIGHT
        elif node.value < value:
            node.right =  self.remove(node.right, value)
        # MATCH FOUND
        elif node.value == value:
            # Case-1: 2 children
            if node.left and node.right:
                leftSubTreeMax = self.max(node.left)
                node.value = leftSubTreeMax
                node.left = self.remove(node.left, leftSubTreeMax)
                return node
            # Case-2: Only left child
            elif node.left and not node.right:
                return node.left
            # Case-3: Only right child
            elif not node.left and node.right:
                return node.right
            # Case-4: 0 children
            else:
                return None
        return node

    def max(self, node):
        if node.right:
            return self.max(node.right)
        else:
            return node.value
```
