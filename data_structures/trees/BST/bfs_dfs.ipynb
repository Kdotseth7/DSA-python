```python
def print_tree(root):
    def height(root):
        return 1 + max(height(root.left), height(root.right)) if root else -1  
    nlevels = height(root)
    width =  pow(2,nlevels+1)

    q=[(root,0,width,'c')]
    levels=[]

    while(q):
        node,level,x,align= q.pop(0)
        if node:            
            if len(levels)<=level:
                levels.append([])
        
            levels[level].append([node,level,x,align])
            seg= width//(pow(2,level+1))
            q.append((node.left,level+1,x-seg,'l'))
            q.append((node.right,level+1,x+seg,'r'))

    for i,l in enumerate(levels):
        pre=0
        preline=0
        linestr=''
        pstr=''
        seg= width//(pow(2,i+1))
        for n in l:
            valstr= str(n[0].value)
            if n[3]=='r':
                linestr+=' '*(n[2]-preline-1-seg-seg//2)+ '¯'*(seg +seg//2)+'\\'
                preline = n[2] 
            if n[3]=='l':
               linestr+=' '*(n[2]-preline-1)+'/' + '¯'*(seg+seg//2)  
               preline = n[2] + seg + seg//2
            pstr+=' '*(n[2]-pre-len(valstr))+valstr #correct the potition acording to the number size
            pre = n[2]
        print(linestr)
        print(pstr)   
```

```python
class BST_Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None
        
class BinarySearchTree:
    
    def __init__(self):
        self.root = None
                    
    def insert(self, node, value):
        if not node:
            return BST_Node(value)
        # LEFT
        if node.value > value:
            node.left = self.insert(node.left, value)
        # RIGHT
        elif node.value < value:
            node.right = self.insert(node.right, value)
        return node
    
    # Level Order Traversal: BFS using Queue
    def bfs(self, node):
        queue = list() # To keep track of the level
        traversal_list = list() # To keep track of order of traversal
        queue.append(node)
        while queue:
            current_node = queue.pop(0)
            if current_node.left:
                queue.append(current_node.left)
            if current_node.right:
                queue.append(current_node.right)
            traversal_list.append(current_node.value)
        return traversal_list
    
    def bfs_recursive(self, queue, traversal_list):
        if not queue:
            return traversal_list
        current_node = queue.pop(0)
        if current_node.left:
            queue.append(current_node.left)
        if current_node.right:
            queue.append(current_node.right)
        traversal_list.append(current_node.value)
        return self.bfs_recursive(queue, traversal_list)
            
    # DFS using Stack (Call Stack)
    def dfs_preorder(self, node, traversal_list):
        if not node:
            return None
        
        traversal_list.append(node.value)
        if node.left:
            self.dfs_preorder(node.left, traversal_list)
        if node.right:
            self.dfs_preorder(node.right, traversal_list)
        return traversal_list
    
    def dfs_inorder(self, node, traversal_list):
        if not node:
            return None
        
        if node.left:
            self.dfs_inorder(node.left, traversal_list)
        traversal_list.append(node.value)
        if node.right:
            self.dfs_inorder(node.right, traversal_list)
        return traversal_list
    
    def dfs_postorder(self, node, traversal_list):
        if not node:
            return None
        
        if node.left:
            self.dfs_postorder(node.left, traversal_list)
        if node.right:
            self.dfs_postorder(node.right, traversal_list)
        traversal_list.append(node.value)
        return traversal_list
```

```python
bst = BinarySearchTree()
keys = [40, 30, 35, 22, 21, 28, 39, 33, 80, 20, 48, 90, 45, 70, 98, 101, 60, 55, 75]
for i in keys:
    bst.root = bst.insert(bst.root, i)
print('-------------------------------------------------------PRINT TREE---------------------------------------------------------------------------------------')
print_tree(bst.root)
print('-------------------------------------------------------DFS PREORDER---------------------------------------------------------------------------------------')
preorder = bst.dfs_preorder(bst.root, list())
print(preorder)
print('-------------------------------------------------------DFS INORDER---------------------------------------------------------------------------------------')
inorder = bst.dfs_inorder(bst.root, list())
print(inorder)
print('-------------------------------------------------------DFS POSTORDER---------------------------------------------------------------------------------------')
postorder = bst.dfs_postorder(bst.root, list())
print(postorder)
print('-------------------------------------------------------BFS----------------------------------------------------------------------------------------')
bfs = bst.bfs(bst.root)
print(bfs)
print('-------------------------------------------------------BFS RECURSIVE----------------------------------------------------------------------------------------')
bfs_recursive = bst.bfs_recursive([bst.root], list())
print(bfs_recursive)
```
