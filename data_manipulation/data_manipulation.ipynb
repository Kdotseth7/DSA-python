```python
# Column String operations
# Find and Replace
df['geo_desc'] = df['geo_desc'].str.replace('States', 'god')
df.head()

# Lower
df['geo_desc'] = df['geo_desc'].str.lower()
df.head()

# Len
df['len_geo_desc'] = df['geo_desc'].str.len()
df.head()
```

```python
# Dropping Rows based on conditions
df = df[df['cat_code'] != 'SOLD']
df = df[(df['cat_code'] != 'SOLD') & (df['dt_idx'] != 1)]
df.head()
```

```python
# Select rows using isin() clause
df = df[df['cat_idx'].isin([1, 3])]
df.head()
```

```python
# Lambda Apply function pandas
df['dt_idx_sq'] = df['dt_idx'].apply(lambda x: x**2)
df['tokenized_desc'] = df['cat_desc'].apply(lambda x: x.split())
df.head()
```

```python
# Apply function on a single row of multiple columns
def random_fn(row2, row3):
    return row2 == 1 and row3 == 'SOLD'

df['random_op'] = df.apply(lambda row: random_fn(row[2], row[3]), axis=1)
df.head()
```

```python
# Rename columns
df.rename(columns={
    'cat_code': 'dog_code',
    'cat_desc': 'dog_desc'
}, inplace=True)
df.head()
```

```python
# Create DataFrame for employees
emp = {
    'emp_id': [1, 2, 3, 4, 5, 6, 7, 8],
    'emp_name': ['Jack', 'Rob', 'Moe', 'Kevin', 'Paul', 'Stephen', 'Jeff', 'Brendan'], 
    'manager_id': [7, 5, 3, 2, 1, 1, 2, 1]
}
emp_df = pd.DataFrame.from_dict(emp)
emp_df
```

```python
# SELF JOIN -> To Find list of Employees and their Managers
merged_df = emp_df.merge(emp_df, how='inner', left_on='emp_id', right_on='manager_id')
merged_df = merged_df[['emp_name_x', 'emp_name_y']]
merged_df.rename(columns={
    'emp_name_x': 'emp', 
    'emp_name_y': 'manager'
}, inplace=True)
merged_df
```

```python
# Create Customer Table
customers = {
    'id': [1, 2, 3, 4, 5, 6, 7, 8],
    'name': ['Jack', 'Rob', 'Moe', 'Kevin', 'Paul', 'Stephen', 'Jeff', 'Brendan'], 
    'country': ['Ind','USA', 'USA', 'Iran', 'Iraq', 'Korea', 'India', 'Pakistan']
}
cust_df = pd.DataFrame.from_dict(customers)
cust_df
```

```python
# Create Order Table
orders = {
    'order_id': [101, 202, 3402, 4222, 512, 622, 711, 811],
    'customer_id': [1, 2, 1, 5, 5, 1, 3, 6],
}
orders_df = pd.DataFrame.from_dict(orders)
orders_df
```

```python
# Left Join customers and orders tables
merged_data = cust_df.merge(orders_df, how='left', left_on='id', right_on='customer_id')
merged_data
# Drop NaN rows
merged_data.dropna(axis=0, inplace=True)
merged_data
```
